# MULTI-AGENTIC RAG WITH HUGGING FACE CODE AGENTS


This notebook accompanies the blog post Multi-agentic RAG with Hugging Face Code Agents. The full code implementation can be found in the GitHub repo https://github.com/GabrieleSgroi/multiagentic_rag.

Code agents work by executing Python code, make sure you run the notebook in an isolated environment and don't allow any unsafe import or function. Always supervise the system execution inspecting the logs, interrupt the execution if the system is trying to execute code that may pose risks.

As all Large Language Model based systems, this is also prone to hallucination. Always check the sources used to generate the answers and don't trust the generated answers blindly.

By default, the agents are powered by the model *Qwen2.5–7B-Instruct*. See the [model card](https://huggingface.co/Qwen/Qwen2.5-7B-Instruct) for licence, limitations, and other information.

The default embedding model for similarity retrieval is all-mpnet-base-v2, check also the corresponding [model card](https://huggingface.co/sentence-transformers/all-mpnet-base-v2
).

```python
from transformers import BitsAndBytesConfig, AutoTokenizer, AutoModelForCausalLM
from multirag.multiagentic_system import MultiAgenticRAG
import torch
```

# Load model


```python
nf4_config = BitsAndBytesConfig(load_in_4bit=True,
                                bnb_4bit_quant_type="nf4",
                                bnb_4bit_use_double_quant=True,
                                bnb_4bit_compute_dtype=torch.bfloat16)
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-7B-Instruct",
                                          trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen2.5-7B-Instruct",
                                              device_map="cuda",
                                              torch_dtype=torch.float16,
                                              trust_remote_code=True,
                                              quantization_config=nf4_config)
```

```python
multiagentic_rag = MultiAgenticRAG(model=model, tokenizer=tokenizer)
```

# Multi-agentic RAG

```python
answer = multiagentic_rag("Does birch plywood float in ethanol?")
```

```python
answer
```

## Manager agent logs

```python
for s in multiagentic_rag.manager_agent_logs['steps']:
  print("AGENT: ", s['llm_output'])
  print("OBSERVATION: ", s['observation'])
```

## Wikipedia search agent logs

```python
for r in multiagentic_rag.wikipedia_search_agent_logs:
  print("TASK: ", r['task'])
  for s in r['steps']:
    print('AGENT: ', s['llm_output'])
    print('OBSERVATION: ', s['observation'])
  print("---"*10)
```
